# Pré processamento dos dados


```python
import numpy as np
import pandas as pd

dados = pd.read_csv('../tratado.csv') 

dados.head()
dados.shape
```

## Tranformação das variaveis categoricas nominais

```python
# fazer uma copia do dataset, para trabalhar sem riscos
dados2 = pd.DataFrame.copy(dados)
```

```python
dados2.head()
```

```python
# fazer alterações nas colunas de nominais para ordinais
dados2['Sex'].replace({'M': 0, 'F': 1}, inplace=True)
dados2['ChestPainType'].replace({'TA': 0, 'ATA': 1, 'NAP': 2, 'ASY': 3,}, inplace=True)
dados2['RestingECG'].replace({'Normal': 0, 'ST': 1, 'LVH': 2}, inplace=True)
dados2['ExerciseAngina'].replace({'N': 0, 'Y': 1}, inplace=True)
dados2['ST_Slope'].replace({'Up': 0, 'Flat': 1, 'Down': 2}, inplace=True)
```

```python
dados2.dtypes
```

```python
# Selecioanr colunas relevantes para classificação
previsores = dados2.iloc[:, 0:11].values
```

```python
previsores
```

```python
previsores.shape
```

```python
#meu alvo == target
alvo = dados.iloc[:, 11].values
```

```python
alvo
```

```python
# usar padronização ou normalização para ajustar desvios absurdos
# Padronização (utilizar média e descio como ref.)
#  Normalização (utilizar os valores máximo e mínimo como ref.)

from sklearn.preprocessing import StandardScaler # padronização
```

### labelEncoder

```python
# trnasformação de var. categoricas em númericos de forma automatica.
from sklearn.preprocessing import LabelEncoder
```

```python
dados.head()
```

```python
previsores2 = dados.iloc[:,0:11].values
previsores2
```

```python
previsores2[:,1] = LabelEncoder().fit_transform(previsores2[:,1]) #seleciona o indice da coluna e transforma
previsores2[:,10] = LabelEncoder().fit_transform(previsores2[:,10])
previsores2

```

### OneHotEncoder

```python
# criação de dummys
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ColumnTransformer
```

```python
#aplicar transformação
previsores3 = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1,2,6,8,10])], remainder='passthrough').fit_transform(previsores2)
```

```python
previsores3
```

```python
previsores3 = pd.DataFrame(previsores3)
previsores3.head()
```

```python
# vai transformar previsores com relação a sua scscala
previsores3_esc = StandardScaler().fit_transform(previsores3)
```

```python
previsores3_esc
```

```python
dados4 = pd.DataFrame(previsores3_esc)
dados4
```

```python
dados4.describe()
```

## **LEGENDA**

Age = idade (anos)

Sex = sexo (0=M; 1=F)

Chest Pain Type = tipo de dor no peito (0=TA:angina típica; 1=ATA:angina atípica; 2=NAP:dor não anginosa; 3=ASY:assintomático)

Resting BP = pressão sanguínea em repouso (mmHg)

Cholesterol = colesterol sérico (mg/dl)

Fasting BS = açúcar no sangue em jejum (mg/dl)
          0: Fasting BS < 120 mg/dl (não diabético)
          1: Fasting BS >= 120 mg/dl, (diabético)

Resting ECG = eletrocardiograma em repouso (0=Normal; 1=ST:Anormalidade da onda ST-T; 2=LVH: Hipertrofia ventricular esquerda)

Max HR = frequência cardíaca máxima

Exercise Angina = Angina induzida por exercício (0=Não; 1=Sim)

Old Peak = Depressão de ST induzida por exercício em relação ao repouso

ST_Slope = Inclinação do segmento ST (0=UP; 1=Flat; 2=Down)

Heart Disease = Doença cardíaca (0=Não possui doença cardíaca; 1=Possui doença cardíaca)

### descrição

alvo = variável que se pretende atingir (tem ou não doença cardíaca).

previsores = conjunto de variáveis previsoras com as variáveis categóricas transformadas em numéricas manualmente, sem escalonar.

previsores_esc = conjunto de variáveis previsoras com as variáveis categóricas transformadas em numéricas, escalonada.

previsores2 = conjunto de variáveis previsoras com as variáveis categóricas transformadas em numéricas pelo labelencoder.

previsores3 = conjunto de variáveis previsoras transformadas pelo labelencoder e onehotencoder, sem escalonar.

previsores3_esc = conjunto de variáveis previsoras transformadas pelo labelencoder e onehotencoder escalonada.
